```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataframe=pd.read_csv('Zomato data .csv')
```

```python
dataframe.head()
```

```python
dataframe.isnull().sum()
```

```python
## The rate coloum has that float values that like 4.1/5 so the 5 is distrube that so we remove [5].

def handleRate(value):
    value=str(value).split('/')
    value=value[0];
    return float(value)

dataframe['rate']=dataframe['rate'].apply(handleRate)

    
```

```python
dataframe.head()
```

```python
dataframe.info()
```

```python
dataframe.describe()
```

```python
# Types of resturants

sns.countplot(x=dataframe['listed_in(type)'])
```

```python
 dataframe.head()
```

```python
# how many votes each type resuturant recieved from customers

g_data = dataframe.groupby('listed_in(type)')['votes'].sum()
result = pd.DataFrame({'Votes': g_data})

# Change plt.Plot to plt.plot
plt.plot(result, c="black",marker='o') 

plt.xlabel('types of restaurant', c='green', size=20)
plt.ylabel('votes', c='red')

# Add plt.show() to display the plot
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'dataframe' is already defined and has the necessary columns

pivot_table = dataframe.pivot_table(index='listed_in(type)', columns='online_order', aggfunc='size', fill_value=0)

# Use a valid colormap name, e.g., 'YlGnBu'
sns.heatmap(pivot_table, annot=True, cmap='YlGnBu', fmt='d')

plt.title('Heatmap of Online Orders by Listing Type')
plt.xlabel('Online Order')
plt.ylabel('Listed In Type')
plt.show()
```
